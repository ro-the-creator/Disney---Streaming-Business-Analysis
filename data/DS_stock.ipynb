```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/Marcy_Student/Desktop/m0-Final_Project/data/DIS.US_H4.csv")
```

```python
# Creating a filtered_df between Nov 12, 2019 - Nov 12, 2022

filtered_nov_df = pd.Series({
    'datetime': pd.date_range(start='2019-11-12 20:00:00', end='2022-11-12 20:00:00')
    })

filtered_df = df.loc[df['datetime'] > '2019-11-12 20:00:00']

print(filtered_df)
```

```python
# changing columns to int

filtered_df['close'] = filtered_df['close'].astype(int)
filtered_df['open'] = filtered_df['open'].astype(int)
filtered_df['high'] = filtered_df['high'].astype(int)
filtered_df['low'] = filtered_df['low'].astype(int)
filtered_df.dtypes
```

```python
fdf_mean = np.mean(filtered_df["close"])
fdf_median = np.median(filtered_df["close"])
fdf_std = np.std(filtered_df["close"])
Q1 = np.percentile(filtered_df['close'], 25)
Q3 = np.percentile(filtered_df['close'], 75)
IQR = np.percentile(filtered_df['close'], 75) - np.percentile(filtered_df['close'], 25)

print("Closing Stock Price Analysis From Nov 12, 2019 - Nov 12, 2022")
print("Media :"," $", fdf_mean)
print("Mean :"," $", fdf_median)
print("---------------------------")
print("Strd Dev. :"," $", fdf_std)
print("Q1 :", "$", Q1)
print("Q3 :", "$", Q3)
print("IQR :", "$", IQR)
```

```python
Labels = ['2019', '2020', '2021', '2022']

plt.plot(filtered_df["datetime"], filtered_df['close'])
plt.xlabel("Date (Yr)")
plt.ylabel("Closing Value ($)")
plt.title("Disney Closing Stock Value Nov 2019 - Nov 2022")
plt.grid(True)
# plt.xticks(filtered_df['datetime'], Labels)
plt.xticks([200, 1040, 1840], ['2020', '2021', '2022'])
plt.show() # Displays the graph
```
